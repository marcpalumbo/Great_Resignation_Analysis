```python
#Drop rows where year column has data with 'D' in it. Making a copy to not interfere with the new DataFrame.
df2= new_df

# Fiscal Year 2017(Quarter 1 - Quarter4)
df2.drop(df2[df2['2017:Q1'] == '(D)'].index,inplace = True)
df2.drop(df2[df2['2017:Q2'] == '(D)'].index,inplace = True)
df2.drop(df2[df2['2017:Q3'] == '(D)'].index,inplace = True)
df2.drop(df2[df2['2017:Q4'] == '(D)'].index,inplace = True)

# Fiscal Year 2018(Quarter 1 - Quarter4)
df2.drop(df2[df2['2018:Q1'] == '(D)'].index,inplace = True)
df2.drop(df2[df2['2018:Q2'] == '(D)'].index,inplace = True)
df2.drop(df2[df2['2018:Q3'] == '(D)'].index,inplace = True)
df2.drop(df2[df2['2018:Q4'] == '(D)'].index,inplace = True)

# Fiscal Year 2019(Quarter 1 - Quarter4)
df2.drop(df2[df2['2019:Q1'] == '(D)'].index,inplace = True)
df2.drop(df2[df2['2019:Q2'] == '(D)'].index,inplace = True)
df2.drop(df2[df2['2019:Q3'] == '(D)'].index,inplace = True)
df2.drop(df2[df2['2019:Q4'] == '(D)'].index,inplace = True)

# Fiscal Year 2020(Quarter 1 - Quarter4)
df2.drop(df2[df2['2020:Q1'] == '(D)'].index,inplace = True)
df2.drop(df2[df2['2020:Q2'] == '(D)'].index,inplace = True)
df2.drop(df2[df2['2020:Q3'] == '(D)'].index,inplace = True)
df2.drop(df2[df2['2020:Q4'] == '(D)'].index,inplace = True)
```
