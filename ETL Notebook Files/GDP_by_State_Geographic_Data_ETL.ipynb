```python
# Created a new list of names of the 'Description' column and recreated it with a '- ' partial string to then be identified later to be dropped.
# This is for the reason that those chosen rows are total values and will skew the data. 
# In addition some chosen rows are subsets of an industry as well as oversees activity in which we are focused only on activities within the U.S.
old_keys = [
"All industry total",
"  Private industries",
"      Agriculture, forestry, fishing and hunting",
"      Mining, quarrying, and oil and gas extraction",
"      Utilities",
"      Construction",
"      Manufacturing",
"          Durable goods manufacturing",
"          Nondurable goods manufacturing",
"      Wholesale trade",
"      Retail trade",
"      Transportation and warehousing",
"      Information",
"      Finance and insurance",
"      Real estate and rental and leasing",
"      Professional, scientific, and technical services",
"      Management of companies and enterprises",
"      Administrative and support and waste management and remediation services",
"      Educational services",
"      Health care and social assistance",
"      Arts, entertainment, and recreation",
"      Accommodation and food services",
"      Other services (except government and government enterprises)",
"  Government and government enterprises",
"    Federal civilian",
"    Military",
"    State and local",
]

new_keys = [
"- All industry total",
"- Private industries",
"Private industries Agriculture, forestry, fishing and hunting",
"Private industries Mining, quarrying, and oil and gas extraction",
"Private industries Utilities",
"Private industries Construction",
"- Manufacturing",
"Private industries Manufacturing Durable goods manufacturing",
"Private industries Manufacturing Nondurable goods manufacturing",
"Private industries Wholesale trade",
"Private industries Retail trade",
"Private industries Transportation and warehousing",
"Private industries Information",
"Private industries Finance and insurance",
"Private industries Real estate and rental and leasing",
"Private industries Professional, scientific, and technical services",
"Private industries Management of companies and enterprises",
"Private industries Administrative and support and waste management and remediation services",
"Private industries Educational services",
"Private industries Health care and social assistance",
"Private industries Arts, entertainment, and recreation",
"Private industries Accommodation and food services",
"Private industries Other services (except government and government enterprises)",
"- Government and government enterprises",
"Government and government enterprises Federal civilian",
"Government and government enterprises Military",
"Government and government enterprises State and local ", 
    
]
```

```python
# For loop statements to iterate through the 'Description' column and identify the partial string '- ' to then be replaced by '0'.
new_keys2 = []
for i in range(len(new_keys)):
    if new_keys[i][0:2] != '- ':
        new_keys2.append(new_keys[i])
    else:
        new_keys2.append('0')
        
new_keys2
```

```python
repl = {x:y for [x, y ] in zip(old_keys, new_keys2)}
```

```python
df3 = df3.replace(repl)
```

```python
df4 = df3[df3['Description'] != '0']
df4.head()
```
